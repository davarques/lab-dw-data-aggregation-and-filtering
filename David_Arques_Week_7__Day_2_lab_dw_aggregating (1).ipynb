```python
# Reading csv file:
url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv'
df = pd.read_csv(url)
df = df.copy() # safety copy
df.head()
```

#### Task 1:
Create a new DataFrame that only includes customers who have a total_claim_amount greater than $1,000 and have a response of "Yes" to the last marketing campaign.

```python
df_task_1 = df[(df['Total Claim Amount'] > 1000) & (df['Response'] == 'Yes')]
df_task_1
```

#### Task 2:
Using the original Dataframe, analyze the average total_claim_amount by each policy type and gender for customers who have responded "Yes" to the last marketing campaign. Write your conclusions.

```python
# First pulling a df for customers who have responded "Yes" in ['Response']

df_response_yes = df[df['Response'] == 'Yes']
df_response_yes
```

```python
# Grouping by 'Policy Type' and 'Gender'
policy_gender_groupby = df_response_yes.groupby(['Policy Type', 'Gender'])
policy_gender_groupby.head(1)
```

```python
# Calculate the mean of 'Total Claim Amount' for each group
avg_claim_amount_by_policy_gender = policy_gender_groupby['Total Claim Amount'].mean()
avg_claim_amount_by_policy_gender
```

#### Task 3:
Analyze the total number of customers who have policies in each state, and then filter the results to only include states where there are more than 500 customers.

```python
# First selecting customers that have policies (so at least 1):
customers_with_policies = df[df['Number of Policies'] >= 1]
```

```python
# Grouping by 'State' and count the number of customers in each state
total_customer_policies_state = customers_with_policies.groupby('State')['Customer'].count()
total_customer_policies_state
```

```python
# Filtering only states where there are more than 500 customers:
states_with_only_500_customers = total_customer_policies_state[total_customer_policies_state > 500]
states_with_only_500_customers
```

Find the maximum, minimum, and median customer lifetime value by education level and gender. Write your conclusions.
